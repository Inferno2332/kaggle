```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tl-data.csv')
df2 = df[df['40l_sprint'].notna()]
df2.head()
```

```python
def percentile(n):
    def percentile_(x):
        return round(np.percentile(x, n), 3)
    if n <= 50:
        percentile_.__name__ = f'Fastest {n}%'
    else:
        percentile_.__name__ = f'Slowest {100-n}%'
    return percentile_

def mean(x):
    return round(np.mean(x), 3)
```

```python
time_stats = df2.groupby('rank')['40l_sprint'].agg([percentile(5), 
                                                    percentile(25), 
                                                    'median', 
                                                    percentile(75), 
                                                    percentile(95)]).reset_index()
```

```python
# Order the ranks properly

df_mapping = pd.DataFrame({
    'rank': ['d', 'd+', 'c-', 'c', 'c+', 'b-', 'b', 'b+', 'a-', 'a', 'a+', 's-', 's', 's+', 'ss', 'u', 'x'],
})

sort_mapping = df_mapping.reset_index().set_index('rank')

time_stats['rank_num'] = time_stats['rank'].map(sort_mapping['index'])

time_stats.sort_values('rank_num', inplace=True)
time_stats.drop('rank_num', axis=1, inplace=True)
time_stats['rank'] = time_stats['rank'].apply(str.upper)

time_stats
```

```python
time_stats = df2.groupby('rank')['blitz'].agg([percentile(5), 
                                                    percentile(25), 
                                                    'median', 
                                                    percentile(75), 
                                                    percentile(95)]).reset_index()

# Order the ranks properly

df_mapping = pd.DataFrame({
    'rank': ['d', 'd+', 'c-', 'c', 'c+', 'b-', 'b', 'b+', 'a-', 'a', 'a+', 's-', 's', 's+', 'ss', 'u', 'x'],
})

sort_mapping = df_mapping.reset_index().set_index('rank')

time_stats['rank_num'] = time_stats['rank'].map(sort_mapping['index'])

time_stats.sort_values('rank_num', inplace=True)
time_stats.drop('rank_num', axis=1, inplace=True)
time_stats['rank'] = time_stats['rank'].apply(str.upper)

time_stats
```

```python
r = input('Enter rank: ')
all_times = df2['40l_sprint']
times = df2[df2['rank'] == r]['40l_sprint']

t = float(input('Enter 40L time: '))

pctl = sum(times > t) / len(times)
pctl_all = sum(all_times > t) / len(all_times)

print(f'\nYour sprint is faster than {100*pctl:.2f}% of people at your rank.')
print(f'It is faster than {100*pctl_all:.2f}% of ranked players.')
```
